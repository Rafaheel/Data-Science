# Filtrando DataFrames

```python
#vamos criar uma lista para saber os filmes que durão mais que 180 minutos.

boleanos = []

for length in filmes.duration:
    if length >= 200:
        boleanos.append(True)
    else:
        boleanos.append(False)
```

```python
boleanos[:20]
```

```python
# é possivel transformar uma lista em serie
duração = pd.Series(boleanos)
type(duração)
```

```python
#Para series boleanos ou para indexar e filtrar, é preciso usar a notação de colchetes
print(filmes['title'] [:10])
```

```python
#aplicando a serie no data frame inteiro
filmes[duração]
```

```python
# Aplicando usando apenas as colunas desejadas.
filmes[duração] [['title','star_rating']]
```
